```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from sklearn import datasets
data = datasets.load_boston()
data
```

```python
all_data = pd.DataFrame(data.data, columns=data.feature_names)
all_data
```

```python
all_data['MEDV'] = data.target
all_data.head()
```

```python
all_data.describe()
```

```python
fig, axs = plt.subplots(nrows=5, ncols=3, figsize = (20,10))
index = 0
axs = axs.flatten()
for col in all_data.columns:
    sns.histplot(all_data[col], ax=axs[index], kde=True)
    index += 1
plt.tight_layout()
```

```python
all_data2 = all_data.drop(columns = ["RAD"], axis=1)
all_data2
```

```python
all_data2['ZN'].unique()
```

```python
from sklearn.preprocessing import RobustScaler

cols = [col for col in all_data2.iloc[:,:-1].columns if col not in ['CHAS']]
rs = RobustScaler()
all_data2[cols] = rs.fit_transform(all_data2[cols])
all_data2
```

```python
fig = plt.figure(figsize=(20, 10))
sns.boxplot(data=all_data2.iloc[:,:-1])
```

```python
from sklearn.model_selection import train_test_split
x_scaled = all_data2.iloc[:,:-1]
y = all_data2['MEDV']
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 13)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
predict_lr = lr.predict(x_test)
lr_mse = mean_squared_error(y_test, predict_lr)
lr_rmse = np.sqrt(lr_mse)
print(lr_mse, lr_rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
rf.score(x_train, y_train)
```

```python
rf.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
predict_rf = rf.predict(x_test)
rf_mse = mean_squared_error(y_test, predict_rf)
rf_rmse = np.sqrt(rf_mse)
print(rf_mse, rf_rmse)
```

```python
from sklearn.linear_model import RidgeCV
alpha_list = np.random.uniform(0,10,100)
rd = RidgeCV(alphas=alpha_list)
rd.fit(x_train, y_train)
```

```python
alpha = rd.alpha_
alpha
```

```python
from sklearn.linear_model import Ridge
rd = Ridge(alpha=alpha)
rd.fit(x_train, y_train)
```

```python
rd.score(x_train, y_train)
```

```python
rd.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
predict_rd = rd.predict(x_test)
rd_mse = mean_squared_error(y_test, predict_rd)
rd_rmse = np.sqrt(rd_mse)
print(rd_mse, rd_rmse)
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import keras
NB_EPOCHS = 500
BATCH_SIZE = 50
learning_rate=0.001

model = Sequential()
model.add(Dense(12, activation='ReLU', input_shape=(12,)))
model.add(Dense(12, activation='ReLU'))
model.add(Dense(1))

model.compile(optimizer=tensorflow.keras.optimizers.Adam(learning_rate=learning_rate), loss='mse', metrics=['mse'])
history = model.fit(x_train, y_train, validation_data=(x_test,y_test), epochs=1000, verbose=2, batch_size=BATCH_SIZE, 
                    callbacks = [keras.callbacks.EarlyStopping(patience=100)])
```

```python
def modelplot(history, data, val_data):
  plt.plot(history.history[data])
  plt.plot(history.history[val_data])
  plt.title('Model'+ data)
  plt.ylabel(data)
  plt.xlabel('epoch')
  plt.legend(['train', 'test'])
  plt.show()
```

```python
modelplot(history, 'loss', 'val_loss')
```
